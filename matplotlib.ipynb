```python
import matplotlib.pyplot as plt
```

```python
#Defining a range to draw simple line
plt.plot(range(10))
```

```python
year=[1950,1970,1990,2000]
pop=[2.15,3.07,3.93,4.09]
plt.scatter(year,pop)
```

```python
from matplotlib import style
plt.plot(year,pop)
plt.title("World population")
plt.xlabel("years")
plt.ylabel("pop")
plt.style.use('ggplot')
```

```python
values=[12,55,4,32,14]
colors=['r','g','b','c','m']
explode=[0.2,0,0.5,0,0]
labels=["India","United States",'Russia',"China","Europe"]
```

```python
plt.pie(values,colors=colors,labels=labels,explode=explode,shadow=True,autopct='%1.3f%%',counterclock=True)
plt.title('Student Locations')
```
