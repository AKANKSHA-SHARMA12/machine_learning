```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs
```

```python
data=make_blobs(n_samples=200,n_features=2,random_state=101)
```

```python
plt.scatter(data[0][:,0],data[0][:,1],cmap="rainbow")
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
f, (ax1,ax2) = plt.subplots(1,2, sharey= True, figsize = (10,6))
ax1.set_title("K Means")
ax1.scatter(data[0][:,0], data[0][:,1], c = kmeans.labels_, cmap = "rainbow")
ax2.set_title("original")
ax2.scatter(data[0][:,0], data[0][:,1], c= data[1],cmap = "rainbow")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_blobs
x,y=make_blobs(n_samples = 300, centers = 5, 
                  cluster_std = 0.6)
```

```python
x.shape
```

```python
y.shape[0]
```

```python
from sklearn.cluster import KMeans
```

```python
WCV = []
```

```python
for i in range(1, 16):
    km = KMeans(n_clusters = i) # Yeh 15 objects banayega KMeans class ka. Jisme K ki value 1 se lekar 15 tak hogi
    km.fit(x)
    WCV.append(km.inertia_)
```

```python
WCV
```

```python

plt.plot(range(1, 16), WCV)
plt.show()
```

```python
km = KMeans(n_clusters = 5)
y_pred = km.fit_predict(x)
```

```python
plt.scatter(x[y_pred == 0, 0], x[y_pred == 0, 1])
plt.scatter(x[y_pred == 1, 0], x[y_pred == 1, 1])
plt.scatter(x[y_pred == 2, 0], x[y_pred == 2, 1])
plt.scatter(x[y_pred == 3, 0], x[y_pred == 3, 1])
plt.scatter(x[y_pred == 4, 0], x[y_pred == 4, 1])
plt.show()
```
